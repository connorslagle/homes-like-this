```python
# general imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# image imports
from skimage import io
```

```python
df = pd.read_csv('data/metadata/2020-06-09_pg1_3_all.csv')
df.info()
```

```python
df.head()
```

```python
df['image_file'][0]
```

```python
def imgs_in_df(df, dir_path = 'data/proc_images/color/128'):
    imgs_dct = {'fpath': [], 'img_obj': []}
    not_in_path = {'fpath': [], 'index': []}
    for file in df['image_file']:
        try:
            fpath = f'data/proc_images/color/128/{file}'
            img = io.imread(fpath)
            imgs_dct['fpath'].append(file)
            imgs_dct['img_obj'].append(img)
        except:
            not_in_path['fpath'].append(file)
            not_in_path['index'].append()
            continue
    return imgs_dct
        
```

```python
df[df['image_file'] == imgs_dct.get('fpath')[0]]
```

```python
imgs_dct = imgs_in_df(df)
```

```python
plt.close('all')

fig, ax = plt.subplots()
ax.imshow(imgs_dct.get('img_obj')[0])
ax.axis('off')
```
